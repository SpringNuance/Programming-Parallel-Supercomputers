```python
# Performance for MPI baseline

number_of_processes = [2,4,6,8,10,12,16,20,24]

# Extracted data from the provided seff outputs
time_baseline = [83.41, 83.54, 82.743, 82.919, 81.922, 82.608, 83.054, 81.921, 82.445]

time_hybrid = [101.413, 100.65, 100.702, 101.292, 101.287, 101.094, 100.58, 69.672, 69.434]

# Create the plot
plt.figure(figsize=(6, 4))
plt.plot(number_of_processes, time_baseline, marker='o', label='MPI Baseline')
plt.plot(number_of_processes, time_hybrid, marker='s', label='MPI + OpenMP')

# Customize the plot
plt.title('Strong scaling: Time vs Number of Processes', fontsize=14)
plt.xlabel('Number of Processes', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Weak scaling

number_of_processes = [2,4,6,8,10,12,16,20,24]

grid_size = [1000,2000,3000,4000,5000,6000,8000,10000,12000]

time_baseline = [10.846, 21.243, 36.003, 52.103, 69.324, 83.873, 117.638, 154.656, 190.762]

time_hybrid =   [12.435, 23.836, 35.703, 47.316, 56.825, 68.049, 89.766, 109.677, 197.005]

# Create the plot
plt.figure(figsize=(6, 4))
# Plotting both baseline and hybrid times
plt.plot(number_of_processes, time_baseline, label='MPI Baseline', marker='o')
plt.plot(number_of_processes, time_hybrid, label='MPI + OpenMP', marker='s')

# Adding title and labels
plt.title('Weak Scaling: Time vs Number of Processes and Grid Size', fontsize=14)
plt.xlabel('Number of Processes / Grid Size', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)

# Adding grid
plt.grid(True)

# Adding legend
plt.legend()

# Handling ticks to show both processes and grid size
ax = plt.gca()
ax.set_xticks(number_of_processes)
ax.set_xticklabels([f'{n}\n{g}' for n, g in zip(number_of_processes, grid_size)])

# Show plot
plt.show()
```
